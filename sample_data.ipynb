```python
data = pd.read_csv('automotive.txt')
```

```python
data = pd.DataFrame(data)
data
```

```python
data.to_csv("sample_data.csv", 'a')
```

```python
data['time'] = pd.to_datetime(data['time'])
data = data.set_index('time')

# print data columns
for x in data.columns:
    print(x)

# fill in NaNs - forward fill
data.fillna(method='ffill',inplace=True)
# fill in NaNs - backward fill
data.fillna(method='bfill',inplace=True)

# Remove last column with no column heading
del data['Unnamed: 22']

# remove columns that match keywords
for dc in data.columns:
    if ("Average" in dc) or ("(total)" in dc) \
       or ("$" in dc) or ("(mA)" in dc):
        del data[dc]

data.head()
```

```python
data.describe()
#currSpeed	currAcceleration	enginePower	fuelUsed	mileage	tyrePressure	rpmEngine	pedalPosition	airFlowRate	coolantTemperature	driverName	maintainenaceStatus	distanceTravelled
```

```python
data.insert(12,"carFine",[1]*len(data))
```

```python
cutoff_low = data['Calculated instant fuel rate (gal./h)'].quantile(0.05)
cutoff_high = data['Calculated instant fuel rate (gal./h)'].quantile(0.95)

# Set values in col2 to zero if they fall within the top or bottom 5% of col1
data.loc[(data['Calculated instant fuel rate (gal./h)'] < cutoff_low) | (data['Calculated instant fuel rate (gal./h)'] > cutoff_high), 'carFine'] = 0
```

```python
data.loc[(data['Vehicle acceleration (g)'] == 0) | (data['Vehicle speed (mph)'] ==0), 'carFine'] = 0

# data.drop(['Fuel used (gallon)'],axis=1)
```
